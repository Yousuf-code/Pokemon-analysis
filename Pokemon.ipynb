```python
import statsmodels.api as sm

# Ensure there are no NaN values in the columns of interest
df = df.dropna(subset=['height_m', 'weight_kg', 'speed'])

# Define the independent variables (add a constant term to allow statsmodels to fit an intercept)
X = sm.add_constant(df[['height_m', 'weight_kg']])

# Define the dependent variable
y = df['speed']

# Create the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
```

1. Create and print a subset of data called 'df_legends', this subset of data shall only contain legendary Pokemons.  (5pts)

3. Which three non-legendary Pokemon has the best chance to beat Hakamo-o (20pts)?
You want a Pokemon that has the ability based on the follow filtering criteria and filtering sequence:
    1. not legendary
    2. 'against_dragon' value of at least 1.80 AND 'against_fight' value of at least 2.00
    3. Has the 'abilities' of at least of the the following:
    - 'Levitate', 'Telepathy', 'Armor' or 'Hustle' (use regex contains)
    4. Top three Pokemon in 'speed'



```python
df_legends = df[df['is_legendary'] == 1]

# Print the subset
print(df_legends)
```

```python
dx = df_legends.groupby('generation')['hp'].idxmax()

# Retrieve the rows corresponding to these indices
highest_hp_per_gen = df_legends.loc[dx]

# Sort by 'generation' and select only the 'name' and 'hp' columns for display
sorted_hp = highest_hp_per_gen.sort_values(by='generation')[['name', 'hp']]

# Print the result
print(sorted_hp)
```

```python
# Filter out legendary Pokmon
non_legendaries = df[df['is_legendary'] == 0]

# Apply criteria on 'against_dragon' and 'against_fight'
effective_against_hakamo = non_legendaries[(non_legendaries['against_dragon'] >= 1.80) & (non_legendaries['against_fight'] >= 2.00)]

# Filter based on abilities using regex
abilities_regex = r'Levitate|Telepathy|Armor|Hustle'
pokemon_with_abilities = effective_against_hakamo[effective_against_hakamo['abilities'].str.contains(abilities_regex, regex=True)]

# Sort by 'speed' and select top three
top_three_pokemon = pokemon_with_abilities.sort_values(by='speed', ascending=False).head(3)

# Display the top three Pokmon
print(top_three_pokemon[['name', 'speed']])
```
